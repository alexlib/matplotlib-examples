```python
# Topics: line, color, LineCollection, cmap, linecolor
'''
Color parts of a 2D line (curve) with coordinates x and y, based on data in z, creating a LineCollection.
Normally, a colormap and/or norm (way of normalizing the data) will be specified, but by default a 
continuous map will be used.

z can be:
- empty, in which case a default coloring will be used based on the position along the input arrays
- a single number, for a uniform color
- an array of the length of the data, to color according to this data
- an array of a smaller length, in which case the colors are repeated along the curve

Returns the LineCollection created, which can be modified afterwards.

Two examples:
- Coloring with a discrete colormap, based on the slope of the line
- Coloring with a continuous colormap, based on distance (or time) along the path.
    This shows how the linewidth can optionally be modified as a function of the data, and optional transparency (alpha)
    (Compare the streamplot function)
'''

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap, BoundaryNorm


############

def make_segments(x, y):
    '''
    Create list of line segments from x and y coordinates, in the correct format for LineCollection:
    an array of the form   numlines x (points per line) x 2 (x and y) array
    '''

    points = np.array([x, y]).T.reshape(-1, 1, 2)
    segments = np.concatenate([points[:-1], points[1:]], axis=1)
    
    return segments


def make_colored_line(segments, colors, cmap, norm, linewidth=3, **kw ):
    '''Take a list of line segments and render them as a LineCollection'''
    lc = LineCollection(segments, array=colors, cmap=cmap, norm=norm, **kw)
    #lc.set_array(colors)  # where the color data gets set; this can be included as array=colors in the LineCollection constructor 
    lc.set_linewidth(linewidth)
    return lc


def linecolor(x, y, z=None, cmap=plt.get_cmap('copper'), norm=plt.Normalize(0.0, 1.0), linewidth=3, **kw):
    '''
    Plot a colored line with coordinates x and y
    Optionally specify colors in the array z
    Optionally specify a colormap, a norm function and a line width
    '''
    
    # Default colors:
    if z is None:
        z = np.linspace(0.0, 1.0, len(x))
           
    # Special case if a single number
    if not hasattr(z, "__iter__"):  # to check for numerical input -- surely there is a better way
        z = np.array([z])
        
    z = np.asarray(z)
    
   
    segments = make_segments(x, y)
    lc = make_colored_line(segments, z, cmap=cmap, norm=norm, linewidth=linewidth, **kw)
    
    ax = plt.gca()

    #fig, axes = plt.subplots(figsize=(5,5))
    #axes.add_collection(lc)
    ax.add_collection(lc)
    
    return lc
        

    
```

```python
x = np.linspace(0, 4.*np.pi, 1000)
y = np.sin(x)

fig, axes = plt.subplots()

linecolor(x, y)
# This works even though y_deriv is a different length -- wraps round

plt.xlim(x.min(), x.max())
plt.ylim(-1.0, 1.0)
plt.show()
```

```python
x = np.linspace(0, 4.*pi, 1000)
y = sin(x)

fig, axes = plt.subplots()

N = 10
for i in range(N):
    color = i / float(N)
    #print color
    shift = 0.2 * i
    linecolor(x-shift, y, color, cmap="cool")
    
plt.xlim(x.min(), x.max())
plt.ylim(-1.0, 1.0)
plt.show()  
```

```python
################

# Curve colored according to the value of its derivative:

# Logarithmic spacing:
log_x = np.linspace(np.log(0.001), np.log(0.1), 5000)
x = np.exp(log_x)
y = 0.9 * np.sin(1. / x)

# Use finite difference to calculate derivative of y:
y_deriv = (y[1:] - y[:-1]) / (x[1:] - x[:-1]) 

# Create a colormap for red, green and blue:
cmap = ListedColormap(['r', 'g', 'b'])

# Create a 'norm' (normalizing function) which maps data values to the interval [0,1]:
#norm = BoundaryNorm([-1, -0.5, 0.5, 1], cmap.N)  # cmap.N is number of items in the colormap
norm = BoundaryNorm([-1000, -200, 200, 1000], cmap.N)  # cmap.N is number of items in the colormap

fig, axes = plt.subplots()

linecolor(x, y, y_deriv, cmap=cmap, norm=norm, linewidth=100.*x)
# This works even though y_deriv is a different length -- wraps round

plt.xlim(x.min(), x.max())
plt.ylim(-1.0, 1.0)
plt.show()
```

```python
################

# Plot of function in polar coordinates:

max_t = 6*np.pi
theta = np.linspace(0, max_t, 1000)

alpha = 0.3
r = 1 + alpha*cos(7./3. * theta)

x = r * np.sin(theta)
y = r * np.cos(theta)

colors=[0.1]*8+[0.5]*4+[0.8]*2

fig, axes = plt.subplots(figsize=(6,6))

#lc = linecolor(x, y)
linecolor(x, y, colors, alpha=1, cmap='jet', linewidth=10) #, [0.3]*8+[0.4]*4+[0.5]*2)

plt.xlim(x.min() - 0.1, x.max() + 0.1)
plt.ylim(y.min() - 0.1, y.max() + 0.1)

axis('equal')



# Turn off ticks:
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
################

# Non-periodic Lissajous curve

max_t = 15.
t = np.linspace(0, max_t, 5000)
x = np.sin(t)
y = np.sin(t * np.pi)

fig, axes = plt.subplots(figsize=(6,6))

lc = linecolor(x, y, alpha=0.75, linewidth=t, cmap='coolwarm') #plt.get_cmap('brg')) # brg
# We have captured the LineCollection created to modify it later
lc.set_rasterized = True

plt.xlim(x.min() - 0.1, x.max() + 0.1)
plt.ylim(y.min() - 0.1, y.max() + 0.1)
plt.show()

# lc = colored_line(x, y, [0.3]*10+[0.4]*3, alpha=0.75, linewidth=1, cmap=plt.get_cmap('brg')) # brg
```
